```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('finalcleandf.csv')
```

```python
bins = [0, 25000,35000, 45000,60000,np.inf]
#names = ["-25","25-35" ,'35-45', '45-60', '60+']
names = [0,1 ,2, 3, 4,]
df['salaryclass'] = pd.cut(df['salary'], bins, labels=names).astype(int)
```

```python
df.drop(["typeofcontract", 'iscocode','salary', 'descriptionclean'], axis=1, inplace=True)
```

```python
one_hot=pd.get_dummies(df[['province', 'education', 'experience','iscodescription']]).astype(int)
df=df.join(one_hot)
df.drop(['province', 'education', 'experience','iscodescription'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df_target=pd.DataFrame(df['salaryclass'],columns=['salaryclass'])
```

```python
df_feat=df.drop(['salaryclass'], axis=1)
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_feat, df_target, test_size=0.30, random_state=101)
```

```python
from sklearn.svm import LinearSVC
model = LinearSVC(multi_class='ovr')
model.fit(X_train,y_train)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```
